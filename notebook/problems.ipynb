A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
n_w = 10
n_r = 20
n_g = 30

n_balls = n_w + n_r + n_g 

# p(w) = 10 / 60 = 1/6
# p(r) = 20 / 60 = 1/3
# p(g) = 30 / 60 = 1/2

pw = n_w / n_balls
pr = n_r / n_balls
pg = n_g / n_balls


# p(A) = p(3w) = p(w) ∩ p(w) ∩ p(w) = (1/6)**3
# P(B) = P(2r) = p(r) ∩ p(r) = 1/9 
# p(A∪B)=p(A) + p(B) - p(A∩B) 

p_3w = pw**3
p_2r = pr**2


p1 = p_3w + p_2r - (p_3w * p_2r)
print('a1=',p1*100,'%')
```
